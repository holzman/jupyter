```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
f = open('wallclock.txt')
```

```python
x = f.readlines()
```

```python
d = {}
```

```python
for line in x:
    (i, w) = line.split(' ')
    if not d.get(i): d[i] = []
    d[i].append(float(w))
```

```python
for x in d.keys(): 
    plt.hist(d[x], 50, label=x)
    plt.xlabel('Committed Slot Time (s)')
    plt.title("Wallclock for mu2e job #15")
    plt.legend()
```
